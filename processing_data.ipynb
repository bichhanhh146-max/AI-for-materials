```python
import pandas as pd
import re

# Đọc file gốc df2, KHÔNG đọc clean_data nếu file đó đã rỗng
df_raw = pd.read_csv(r"C:\Users\Admin\Downloads\20220801_rawdata\data_semiconductor.csv")

# Chỉ giữ 4 cột cần thiết
cols = ["composition", "Temperature", "Thermal conductivity", "DOI"]
df = df[cols].copy()

# Chuẩn hóa rỗng
str_cols = df.select_dtypes(include="object").columns
df[str_cols] = df[str_cols].apply(lambda s: s.str.strip())
df2 = df.replace(r"^\s*$", pd.NA, regex=True)
df2 = df2.dropna(subset=cols).copy()

# Hàm tách nguyên tố
def extract_elements(formula):
    if not isinstance(formula, str):
        return []
    return re.findall(r"[A-Z][a-z]?", formula)

# Lấy tập element từ df1
elements_df1 = set()
for comp in df_raw["Formula"].dropna().astype(str):
    elements_df1.update(extract_elements(comp))

print("Các nguyên tố trong df1:")
print(sorted(elements_df1))
print(f"Tổng số nguyên tố: {len(elements_df1)}")

# Check công thức df2 chỉ chứa element thuộc df1
def is_subset_formula(formula, allowed_elements):
    elems = extract_elements(formula)
    return len(elems) > 0 and all(e in allowed_elements for e in elems)

print(f"\nSố dòng df2 trước khi lọc: {len(df2)}")

mask = df2["composition"].astype(str).apply(lambda x: is_subset_formula(x, elements_df1))
df2_filtered = df2[mask].copy()

print(f"Số dòng df2 sau khi lọc: {len(df2_filtered)}")
print(f"Số dòng bị xóa: {len(df2) - len(df2_filtered)}")

# Lưu lại
df2_filtered.to_csv("clean_empty_&TC_&DOI_&element.csv", index=False)
```

```python
import pandas as pd
import re

# Đọc file
df = pd.read_csv("clean_empty_&TC_&DOI_&element.csv")

# Danh sách nguyên tố hóa học hợp lệ
valid_elements = {
    "H","He","Li","Be","B","C","N","O","F","Ne",
    "Na","Mg","Al","Si","P","S","Cl","Ar","K","Ca",
    "Sc","Ti","V","Cr","Mn","Fe","Co","Ni","Cu","Zn",
    "Ga","Ge","As","Se","Br","Kr","Rb","Sr","Y","Zr",
    "Nb","Mo","Tc","Ru","Rh","Pd","Ag","Cd","In","Sn",
    "Sb","Te","I","Xe","Cs","Ba","La","Ce","Pr","Nd",
    "Pm","Sm","Eu","Gd","Tb","Dy","Ho","Er","Tm","Yb",
    "Lu","Hf","Ta","W","Re","Os","Ir","Pt","Au","Hg",
    "Tl","Pb","Bi","Po","At","Rn","Fr","Ra","Ac","Th",
    "Pa","U","Np","Pu","Am","Cm","Bk","Cf","Es","Fm",
    "Md","No","Lr","Rf","Db","Sg","Bh","Hs","Mt","Ds",
    "Rg","Cn","Nh","Fl","Mc","Lv","Ts","Og"
}

def is_valid_formula(formula):
    # phải là string
    if not isinstance(formula, str):
        return False
    
    # không cho phép space, dấu phẩy, ngoặc, gạch ngang, ký tự lạ
    if formula.strip() != formula:
        return False
    
    # chỉ cho phép chữ cái, số và dấu chấm thập phân
    if not re.fullmatch(r"[A-Za-z0-9.]+", formula):
        return False
    
    # không cho phép công thức bắt đầu/kết thúc bằng dấu chấm
    if formula.startswith(".") or formula.endswith("."):
        return False
    
    # pattern: Element + number
    pattern = r"([A-Z][a-z]?)(\d+(?:\.\d+)?)"
    matches = re.findall(pattern, formula)
    
    # ghép lại phải đúng toàn bộ công thức
    reconstructed = "".join([el + num for el, num in matches])
    if reconstructed != formula:
        return False
    
    # kiểm tra từng nguyên tố
    for el, num in matches:
        if el not in valid_elements:
            return False
        
        # số phải convert được và > 0
        try:
            value = float(num)
            if value <= 0:
                return False
        except:
            return False
    
    return True

# Lọc công thức hợp lệ
df_clean = df[df["composition"].apply(is_valid_formula)].copy()

# Lưu các dòng bị xóa để kiểm tra
df_invalid = df[~df["composition"].apply(is_valid_formula)].copy()

# Reset index
df_clean = df_clean.reset_index(drop=True)
df_invalid = df_invalid.reset_index(drop=True)

# Lưu file
df_clean.to_csv("clean_TC_valid_formula.csv", index=False)
df_invalid.to_csv("invalid_formula_removed.csv", index=False)

print("Số dòng ban đầu:", len(df))
print("Số dòng hợp lệ:", len(df_clean))
print("Số dòng bị xóa:", len(df_invalid))

df_clean.head()
```

```python
import pandas as pd
import re

# Đọc file
df = pd.read_csv("clean_TC_valid_formula.csv")

# Hàm tính tổng thành phần
def get_total(formula):
    try:
        nums = re.findall(r"[A-Z][a-z]?(\d+(?:\.\d+)?)", formula)
        total = sum(float(x) for x in nums)
        return total
    except:
        return None

# Tính tổng
df["total"] = df["composition"].apply(get_total)

# Chia dữ liệu
df_100 = df[abs(df["total"] - 100) < 1e-6].copy()
df_not100 = df[abs(df["total"] - 100) >= 1e-6].copy()

# Reset index
df_100 = df_100.reset_index(drop=True)
df_not100 = df_not100.reset_index(drop=True)

# Lưu file
df_100.to_csv(
    "formula_sum_100.csv",
    index=False
)

df_not100.to_csv(
    "formula_not_sum_100.csv",
    index=False
)

print("Tổng số dòng:", len(df))
print("Số dòng tổng = 100:", len(df_100))
print("Số dòng tổng ≠ 100:", len(df_not100))

df_not100.head()
```

```python
import pandas as pd

# Đọc file
df = pd.read_csv("formula_not_sum_100.csv")

# Chuẩn hóa dữ liệu
df["composition"] = df["composition"].astype(str).str.strip()
df["Temperature"] = pd.to_numeric(df["Temperature"], errors="coerce")
df["Thermal conductivity"] = pd.to_numeric(
    df["Thermal conductivity"],
    errors="coerce"
)

# Group theo composition + Temperature
# TC sẽ được lấy trung bình
df_avg = (
    df.groupby(
        ["composition", "Temperature"],
        as_index=False
    )
    .agg({
        "Thermal conductivity": "mean",
        "DOI": "first"   # giữ DOI đầu tiên
    })
)


# Lưu file mới
output_path = ("TC_averaged.csv")

df_avg.to_csv(output_path, index=False)

print("Số dòng ban đầu:", len(df))
print("Số dòng sau gộp:", len(df_avg))
print("Đã lưu file:", output_path)

df_avg.head()
```
